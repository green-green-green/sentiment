```python
import pandas as pd
```

```python
negative0 = pd.read_csv("negative/0.csv", sep=",")
positive0 = pd.read_csv("positive/0.csv", sep=",")

for i in range(1,10):
    negative = pd.read_csv("negative/{}.csv".format(i))
    positive = pd.read_csv("positive/{}.csv".format(i))
    negative0 = pd.concat([negative0, negative])
    positive0 = pd.concat([positive0, positive])

print(negative0.describe(), positive0.describe())
```

```python
import numpy as np

negativeTweets = negative0["text"]
positiveTweets = positive0["text"]

negaposiDataset = []
for text in negativeTweets:
    negaposiDataset.append([text, 0])

for text in positiveTweets:
    negaposiDataset.append([text, 1])
    
negaposiDatasetJs = []
for line in negaposiDataset:
    dct = {}
    dct["text"] = line[0]
    dct["label"] = line[1]
    negaposiDatasetJs.append(dct)


negaposiDataset = np.array(negaposiDataset)
negaposiDataset
negaposiDatasetJs
```

```python
import csv
with open("negaposiDataset.csv", "w") as w_f:
    writer = csv.writer(w_f)
    writer.writerows(negaposiDataset)
```

```python
import json
with open("negaposiDataset.json", "w") as wjs_f:
    json.dump(negaposiDatasetJs, wjs_f, indent=4)
```
